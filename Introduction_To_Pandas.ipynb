# Combining two columns to generate a new column

```python
area = df1['area']
bathrooms = df1['bathrooms']

# these two columns are still in dataframe format

# converting to a list

areaList = area.values.tolist()
bathroomsList = bathrooms.values.tolist()
```

```python
type(area)
```

```python
type(areaList)
```

```python
spaceList = []

for a,b in zip(areaList,bathroomsList):
    space = a*b
    spaceList.append(space)
```

```python
# creating a new column in df called totalSpacea

df1['totalSpace'] = spaceList
```
